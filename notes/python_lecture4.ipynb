### What are Lists

List is a data type where you can store multiple items under 1 name. More technically, lists act like dynamic arrays which means you can add more items on the fly.

Why Lists are required in programming?<br>
To group related items or data for simplifying or ordering information, condensing your code or performing the same actions on several items at the same time.

## Creating a List

## Deleting items from a List

*   del

```python
# del - delete entire List or elements of it using indexing/slicing
L = [1,2,3,4,5]
print(L)
del L
print(L)
```

## List Functions

*   universal functions - len/min/max/sorted
*   count
*   index
*   reverse
*   sort
*   copy


```python
# len / min / max / sorted
L = [3,7,4,0,6,-4,9]

print(len(L))
print(min(L))
print(max(L))
print(sorted(L))   # ascending order
print(sorted(L,reverse=True))   # descending order
```

```python
L = [1,2,-4,10,'hello']
print(min(L))

# min, max, sorted - supported on homogeneous lists
```

```python
# count
L = [2,6,4,0,2,6,3,0,2,0,-5]

print(L.count(2))
print(L.count(10))
```

```python
# index
L = [2,6,3,0,2,0,-5]

print(L.index(3))
print(L.index(2))    # first occurence
print(L.index(10))
```

```python
# reverse
L = [3,7,4,0,6,-4,9]

# permanently reverses the list
L.reverse()
print(L)
```

```python
# sort vs sorted
L = [3,7,4,0,6,-4,9]
print(L)
print(sorted(L))   # not update original list
print(L)
L.sort()   #  permanently sort the list
print(L)
```

```python
# copy   -> shallow copy
L = [3,7,2,8,0]
print(L)
print(id(L))
L1 = L.copy()    # copy list in new memory location
print(L1)
print(id(L1))
```

## List Comprehension

List Comprehension provides a concise way of creating lists.

newlist = [expression for item in iterable if condition == True]image.png

Advantages of List Comprehension
- More time-efficient and space-efficient than loops.
- Require fewer lines of code.
- Transforms iterative statement into a formula.

```python
# Add 1 to 10 numbers to a list
L = []

for i in range(1,11):
  L.append(i)

print(L)
```

```python
L = [i for i in range(1,11)]
print(L)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = -3
# [-6,-9,-12]

[s*i for i in v]
```

```python
# Add squares
L = [1,2,3,4,5]
[i**2 for i in L]
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50
[i for i in range(1,51) if i%5==0]
```

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']

[i for i in languages if i.startswith('p')]
```

```python
# Nested if with List Comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'
[i for i in my_fruits if i in basket and i.startswith('a')]
```

```python
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
[[(i*j) for i in range(1,4)] for j in range(1,4)]
```

first outer list then inner list<br>
j = 1 then i gets values {1,2,3},
<br> j = 2 then i gets values {1,2,3},
<br> j = 3 then i gets values {1,2,3}

```python
# cartesian products -> List comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]
[i*j for i in L1 for j in L2]
```

```python
[[i*j for i in L2] for j in L1]
```

```python
[[i*j for i in L1] for j in L2]
```

## Zip

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
# zip
L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

z = zip(L1,L2)
print(z,type(z))

L = list(zip(L1,L2))
print(L)

[i*j for i,j in L]
```

```python
# Write a program to add items of 2 lists indexwise
L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]
[i+j for i,j in zip(L1,L2)]
```

#### Create 2 lists from a given list where 1st list will contain all the odd numbers from the original list and the 2nd one will contain all the even numbers

```python
L = [1,2,3,4,5,6]
O = []
E = []
for i in L:
  if i%2 == 0:
    E.append(i)
  else:
    O.append(i)
print('Odd  : {} \nEven : {}'.format(O,E))
```

```python
o = [i for i in L if i%2 == 0]
e = [i for i in L if i%2 != 0]
print('Odd  : {} \nEven : {}'.format(o,e))
```

#### Write a program to replace an item with a different item if found in the list <br>L = [1,2,3,4,5,3] <br>replace 3 with 300

```python
L = [1,2,3,4,5,3]
old_ele = int(input('Enter old element : '))
new_ele = int(input('Enter new element : '))
for i in range(len(L)):
  if L[i] == old_ele:
    L[i] = new_ele
print(L)
```

```python
L = [1,2,3,4,5,3]
old_ele = int(input('Enter old element : '))
new_ele = int(input('Enter new element : '))
L = [i if i != old_ele else new_ele for i in L]
print(L)
```

```python
L = [1,2,3,4,5,3]
old_ele = int(input('Enter old element : '))
if old_ele in L:
  new_ele = int(input('Enter new element : '))
  if L.count(old_ele) > 1:
    L = [i if i != old_ele else new_ele for i in L]
  else:
    L[L.index(old_ele)] = new_ele
else:
  print('Given value is not present')
print(L)
```
